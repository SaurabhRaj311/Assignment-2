#Question 3: An e-commerce platform has customer IDs from two campaigns:

1. Find unique customers across both campaigns.
2. Find customers common in both.
3. Find customers exclusive to campaign1.
4. Explain how sets help in removing duplicates in real DS projects.


**Answers-**

```python
# 1) find unique customers across both campaigns
#Given Data:
campaign1 = {101, 102, 103, 104, 105}
campaign2 = {104, 105, 106, 107}

unique_customers = campaign1.union(campaign2)
print(unique_customers)
```

```python
# 2) find customers common in both campaigns
common_customers = campaign1.intersection(campaign2)
print(common_customers)
```

```python
# 3) find customer exclusive to Campaign 1
exclusive_customers = campaign1.difference(campaign2)
print(exclusive_customers)
```

#Question 5: Given:
sales = [12000, 15000, 11000, 18000, 17000]
Using a loop:
1. Calculate total revenue.
2. Count how many months had sales above 15000.
3. Find maximum sales without using built-in max().


**Answers-**


```python
# 1) calculate total revenue
#given data
sales = [12000, 15000, 11000, 18000, 17000]

total = 0

for i in sales:
    total += i

print("Total Revenue:", total)
```

```python
# 2) count how many months had sales above 15000

count = 0

for i in sales:
    if i > 15000:
        count += 1

print("Months above 15000:", count)
```

```python
# 3) find maximum sales without using built-in max()

maximum = sales[0]

for i in sales:
    if i > maximum:
        maximum = i

print("Maximum Sales:", maximum)
```

#Question 6: You are analyzing product ratings:
ratings = [
[5, 4, 3],

[4, 4, 5],

[3, 5, 4]
]
1. Compute average rating.
2. Print all ratings greater than 4.
3. Explain how nested loops are used in matrix-based ML computations.

```python
# 1) compute average rating
#given data
ratings = [
    [5, 4, 3],
    [4, 4, 5],
    [3, 5, 4]
]



total = 0
count = 0

for row in ratings:
    for value in row:
        total += value
        count += 1

average = total / count
print("Average Rating:", average)
```

```python
# 2) print all ratings greater than 4

for row in ratings:
    for value in row:
        if value > 4:
            print(value)
```

#Question 8: Given
prices = [500, 1500, 2500, 800, 1200]
1. Use map() + lambda to add 18% GST to all prices.
2. Use filter() to select prices above 1000.
3. Explain where such transformations are used in real-world pipelines.


```python
# 1) Use map() + lambda to add 18% GST to all prices
#given
prices = [500, 1500, 2500, 800, 1200]


gst_prices = list(map(lambda x: x + (0.18 * x), prices))
print(gst_prices)
```

```python
# 2) use filters to select prices above 1000

filtered_prices = list(filter(lambda x: x > 1000, prices))
print(filtered_prices)
```

3) Explain where such transformations are used in real-world pipelines

- The functions like map() and filter() are widely used in data science and also in the data engineering pipelines to transform and also to clean the large datasets more efficiently.
For example they are used in applying scaling features, handling of the missing values, taxes and also in selection of relevant records from large datasets. These  operations are very important in preprocessing data before feeding it into the ML models, making sure that the data is consistnet, clean and ready for the analysis.

#Question 9:
Write a generator that yields squares of numbers from 1 to 1,000,000.
Explain difference between:
● List comprehension
● Generator expression

Why are generators preferred in big data scenarios?

**Answers-**

```python
# generator that yields squares of number from 1 to 1,000,000
#generator for squares ( 1 to 1,000,000)
def square_generator():
    for i in range(1, 1000001):
        yield i * i

#using the generator
gen = square_generator()

print(next(gen))
print(next(gen))
print(next(gen))
```

```python
#difference between list comprehension and generator expression

#LIST COMPREHENSION
squares = [x * x for x in range(10)]
# create the entries list in memory at once
# faster for small data
# uses more memory

#GENERATOR EXPRESSION
squares = (x * x for x in range(10))
# do not store full data in memory
#produces the value one by one
# it is more memory efficient
```

 Why generators are preferred in big Data Scenarios?

 - Generators are preferred in big data scenarios because they have the capacity to handle the large datasets more efficeintly without loading everything into memory at once. They generate values on-the-fly, which reduces the memory usage and also improves performance. And this is very useful when working with the huge datasets like streaming data, logs or machine scale learning inputs where storing all data at once is not possible or even practical. Generators make the data processing scalable and efficient in the real-world systems.


#Question 10: Design a class:
class Product:
With:
● Attributes: name, price, quantity
● Method: calculate_revenue()

Create 3 objects and print their revenue.

**Answers-**

```python
# design a class: product with attribute and method
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def calculate_revenue(self):
        return self.price * self.quantity
```

```python
# create 3 objects and print their revenue
p1 = Product("Laptop", 50000, 2)
p2 = Product("Phone", 20000, 3)
p3 = Product("Headphones", 1500, 5)

print("Laptop Revenue:", p1.calculate_revenue())
print("Phone Revenue:", p2.calculate_revenue())
print("Headphones Revenue:", p3.calculate_revenue())
```

Classes are widely used in the building models, pipelines and real world applications where data and behaviour need to handled together efficiently.
